# 需求1: 展示北京一周的天气, 例如: 周一到周日的天气温度变化

```python
# 1. 创建画布.
# figsize: 宽, 高,   dpi: 分辨率(默认是: 100), 越高, 画布越清晰.
# 下述画布像素为: 10 * 80, 5 * 80 -> 800, 400
plt.figure(figsize=(10, 5), dpi=80)

# 2. 准备数据集, x轴, y轴. 
x = [1, 2, 3, 4, 5, 6, 7]       # x轴: 星期.
y = [6, 2, 10, 14, 21, 5, 30]  # y轴: 温度.
plt.plot(x, y)      # plot() -> 绘制 折线图.

# 3. 绘制图像.
plt.show()
```

# 需求2: 画出北京11点到12点1小时内每分钟的温度变化折线图, 温度范围在: 5 ~ 18

```python
# 1. 准备数据. 
# x轴 -> 分钟, 一共 60个值.
x = range(60)
# print(x)

# y轴 -> 温度, 范围: 15 ~ 18, 随机生成.
y = [random.uniform(-5, 18) for i in x]
# print(y)

# 2. 创建画布.
plt.figure(figsize=(20, 8), dpi=80)

# 3. 绘制图形.
plt.plot(x, y)

# 4. 给图标添加辅助功能.
# 4.1 添加自定义x,y刻度
x_ticks_label = [f'11点{i}分' for i in x]      # x的范围: [0, 60),  x轴的: 刻度标签.
y_ticks = [i for i in range(-10, 20)]           # y轴的刻度.

# 具体的设置刻度的操作.
plt.xticks(x[::5], x_ticks_label[::5])       # 0 -> '11点0分',  1 -> '11点1分',  2 -> '11点2分'......
plt.yticks(y_ticks[::5])                     # 0, 1, 2, 3, 4, 5...39

# 4.2 添加网格.
# 参1: 网格线是否显示.  True -> 显示,  False -> 不显示.
# 参2: linestyle: 网格线类型, '-' -> 实线, '--' -> 虚线, ':' -> 点线,  '-.' -> 虚点线
# 参3: 透明度.
plt.grid(True, linestyle='--', alpha=0.8)

# 4.3 设置标题 和 x, y轴的 标签.
plt.xlabel('时间', fontsize=14)
plt.ylabel('温度', fontsize=14)
plt.title('北京11点到12点1小时的温度变化', fontsize=22)

# 5. 保存图形, 细节: plt.show()之后会释放画布资源, 所以要在 plt.show()之前调用, 用于保存图片, 否则图片就是空白的.
# plt.savefig('./data/my2.jpg')

# 6.显示图形.
plt.show()
```

# 需求3: 在同一个画布上, 绘制两个折线. 

```python
# 需求: 画出北京, 漠河, 11点到12点1小时内每分钟的温度变化折线图, 温度范围在: 1 ~ 10

# 1. 准备数据. 
# x轴 -> 分钟, 一共 60个值.
x = range(60)

# y轴 -> 温度, 范围: 15 ~ 18, 随机生成.
y_bj = [random.uniform(5, 18) for i in x]         # y轴 -> 北京.
y_mh = [random.uniform(1, 10) for i in x]         # y轴 -> 漠河.
# print(y)

# 2. 创建画布.
plt.figure(figsize=(20, 8), dpi=80)

# 3. 绘制图形.
plt.plot(x, y_bj, label='北京', color='r')   # label -> 图例标签 显示的值
plt.plot(x, y_mh, label='漠河', color='b')

# 4. 给图标添加辅助功能.
# 4.1 添加自定义x,y刻度
x_ticks_label = [f'11点{i}分' for i in x]      # x的范围: [0, 60),  x轴的: 刻度标签.
y_ticks = [i for i in range(0, 25)]           # y轴的刻度.

# 具体的设置刻度的操作.
plt.xticks(x[::5], x_ticks_label[::5])       # 0 -> '11点0分',  1 -> '11点1分',  2 -> '11点2分'......
plt.yticks(y_ticks[::5])                     # 0, 1, 2, 3, 4, 5...39

# 4.2 添加网格.
# 参1: 网格线是否显示.  True -> 显示,  False -> 不显示.
# 参2: linestyle: 网格线类型, '-' -> 实线, '--' -> 虚线, ':' -> 点线,  '-.' -> 虚点线
# 参3: 透明度.
plt.grid(True, linestyle='-', alpha=0.5)

# 4.3 设置标题 和 x, y轴的 标签.
plt.xlabel('时间', fontsize=15)
plt.ylabel('温度', fontsize=15)
plt.title('北京/漠河 11点到12点1小时的温度变化', fontsize=20)

# 4.4 显示图例.
# 参1: 图例标签,  参2: 图例的位置.
# plt.legend(['北京', '漠河'], loc='best')      # Location Code -> 图例的位置的值.

# 参2: 绘图时如果指定了图例值, 则图例标签可以省略.
plt.legend(loc=0)      # Location Code -> 图例的位置的值,  0 -> best, 自动匹配最佳位置.
# plt.legend(loc='best') # 效果同上.

# 5. 保存图形, 细节: plt.show()之后会释放画布资源, 所以要在 plt.show()之前调用, 用于保存图片, 否则图片就是空白的.
# plt.savefig('./data/my2.jpg')

# 6.显示图形.
plt.show()
```

# 需求4: 面向对象方式绘图 -> 1个画板多个坐标系

```python
# 需求: 画出北京, 漠河, 11点到12点1小时内每分钟的温度变化折线图, 温度范围在: 1 ~ 10 (漠河)    北京: 5 ~ 18 
# 1. 准备数据.  x轴 -> 时间, y轴 -> 温度.
x = range(60)
y_beijing = [random.uniform(5, 18) for i in x]
y_mohe = [random.uniform(1, 10) for i in x]

# 2. 创建画布.
# plt.figure(figsize=(20, 8), dpi=80)       # 面向过程版, 即: 所有的函数都是 plt. 的方式调用的. 
# 面向对象版绘图,  fig -> 画布(1行2列的画布),  ax -> 坐标系
fig, ax = plt.subplots(figsize=(20, 8), nrows=1, ncols=2)

# 3. 具体的绘制图像的动作.
# ax[0] -> 坐标系的0行0列
ax[0].plot(x, y_beijing, color='b', label='北京')
# ax[1] -> 坐标系的0行1列
ax[1].plot(x, y_mohe, color='r', label='漠河', linestyle='--')

# 4. 添加刻度
# 4.1 定义变量, 表示: x轴, y轴刻度的范围. 
x_ticks_label = [f'11点{i}分' for i in x]
y_ticks = [i for i in range(0, 25)]

# 4.2 具体的添加刻度的操作. 
ax[0].set_xticks(x[::5])                    # 图1的: x轴的刻度值
ax[0].set_yticks(y_ticks[::5])              # 图1的: y轴的刻度值
ax[0].set_xticklabels(x_ticks_label[::5])   # 图1的: x轴的(刻度)标签值

ax[1].set_xticks(x[::5])                    # 图2的: x轴
ax[1].set_yticks(y_ticks[::5])              # 图2的: y轴
ax[1].set_xticklabels(x_ticks_label[::5])   # 图2的: x轴的标签值

# 5. 设置标题 和 x, y轴标签.
ax[0].set_xlabel('时间')
ax[0].set_ylabel('温度')
ax[0].set_title('北京11点到12点1小时的温度变化', fontsize=20)

ax[1].set_xlabel('时间')
ax[1].set_ylabel('温度')
ax[1].set_title('漠河11点到12点1小时的温度变化', fontsize=20)

# 6. 添加网格. 
ax[0].grid(True, linestyle='--', alpha=0.5)
ax[1].grid(True, linestyle=':', alpha=0.5)

# 7. 添加图例.
ax[0].legend(loc='best')
ax[1].legend(loc=0)         # 效果同上, 0 -> base, 自动匹配最佳位置.

# 8. 保存图片.
# plt.savefig('./data/my3.jpg')

# 9. 显示图像. 
plt.show()
```

# Matplotlib的其它图形绘制

```python
# 常见图形绘制如下: 折线图( plot() ), 柱状图( bar() ), 直方图( hist() ) 饼图( pie() ), 散点图( scatter() )......
# Matplotlib官方绘图网址 参考: https://matplotlib.org/stable/gallery/index.html
```

```python
# 绘制 柱形图.
# 1. 准备数据.
x = ['A', 'B', 'C', 'D', 'E']
y = [3, 6, 1, 8, 2]

# 创建画布, 如果不写, 系统会默认创建1个画布.
# plt.figure(figsize=(20, 8))

# 2. 绘制柱状图.
plt.bar(x, y, color='r')

# 3. 设置x轴, y轴标签 和 标题.
plt.xlabel('分类名')
plt.ylabel('商品个数')
plt.title('商品分类统计图', fontsize=15)

# 4. 显示图像.
plt.show()
```

```python
# 绘制 直方图 -> 适用于 统计分析, 查看(每组)数据的分布情况. 
# 直方图原理: 相当于把 原数据分成 长度为 区间数 + 1的等差数列, 每两个值 = 1个区间. 
# 柱状图: 展示 单个分类 的值的分布情况. 
# 直方图: 展示 区间值   的分布情况.

# 1. 准备数据.
data = np.random.randint(0, 150, 100)
# data        # 100个 0 ~ 150 之间的随机整数.

# 回顾: 等差数列, 0 ~ 150之间, 11个数.
# print(np.linspace(0, 150, 11))      # [0.  15.  30.  45.  60.  75.  90. 105. 120. 135. 150.]

# 2. 创建画板.
plt.figure(figsize=(20, 8), dpi=80)

# 3. 绘制图形.
# 参1: 绘制的数据. 参2: 直方图的分段数量(即: 几个区间).
# 参3: 颜色, 参4: 透明度, 参5: 柱子的宽度.
plt.hist(data, bins=10, color='b', alpha=0.5, rwidth=0.9)

# 4. 设置 x轴, y轴标签 和 标题.
plt.xlabel('商品价格区间')
plt.ylabel('商品个数')
plt.title('商品价格区间分布图', fontsize=15)

# 5. 显示图形.
plt.show()
```

```python
# 绘制 饼图, 每个分类的占比情况. 
# 1. 参数1: 绘制的数据. 参数2: 饼图的标签.
sizes = [25, 35, 25, 15]
labels = ['电脑', '手机', '键盘', '鼠标']

# 2. 具体的绘图动作. 
# 参1: 绘制的数据. 参2: 饼图的标签. 参3: 显示百分比, %% 表示 1个 %
plt.pie(sizes, labels, autopct='%1.1f%%')  # 绘制饼图，显示百分比
plt.title("Simple Pie Chart")  # 设置图表标题

# 3. 显示图形. 
plt.show()  # 显示图表
```

```python
# 散点图, 适用于 散点分析, 即: 两个值之间的关系. 
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
# scatter(), 散点图.
plt.scatter(x, y, color='red')  # 绘制红色散点图
plt.title("Simple Scatter Plot")  # 设置图表标题
plt.xlabel("X-axis")  # 设置X轴标签
plt.ylabel("Y-axis")  # 设置Y轴标签
plt.show()  # 显示图表
```
