```python
import numpy as np
from numba.core.types import float64
```

### array()

```python
arr1 = np.array([1,2,3])
print(arr1)
print(type(arr1))
print(f'元素类型:{arr1.dtype}')
print(f'元素占用字节数:{arr1.itemsize}')
```

### arange()

```python
arr2 = np.arange(0,10,dtype=int)
print(arr2)
print(type(arr2))
print(f'元素类型:{arr2.dtype}')
print(f'元素占用字节数:{arr2.itemsize}')
```

### 创建随机数
   - np.random.rand(行,列) 生成0.0 - 1.0之间的随机数,包左闭右
   - np.random.randint(起始值, 结束值, size=(行,列)) 生成指定范围内的随机整数
   - np.random.uniform(起始值, 结束值, size=(行,列)) 生成指定范围内的随机小数

```python
arr4 = np.random.rand(3,4)
print(arr4)
print(type(arr4))
print(f'元素类型:{arr4.dtype}')
print(f'元素占用字节数:{arr4.itemsize}')
```

```python
arr5 = np.random.randint(-1,5,size=(3,4))
print(arr5)
print(type(arr5))
print(f'元素类型:{arr5.dtype}')
print(f'元素占用字节数:{arr5.itemsize}')
```

```python
arr6 = np.random.uniform(-1,5,size=(3,4))
print(arr6)
print(type(arr6))
print(f'元素类型:{arr6.dtype}')
print(f'元素占用字节数:{arr6.itemsize}')
```

 ### 类型转换

```python
arr7 = np.zeros((3,4),dtype=np.float64)
print(f'元素类型为:{arr7.dtype}')
```

```python
arr7 = arr7.astype(np.int32)
print(f'元素类型为:{arr7.dtype}')
```
