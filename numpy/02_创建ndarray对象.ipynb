```python
# 导包
import numpy as np
```

# 演示创建ndarray数组的方式

## 1. array()函数

```python
# array()函数 -> 把python中的列表, 封装成 ndarray对象.
arr1 = np.array([1, 2, 3], dtype=np.int32)
print(arr1)
print(type(arr1))
print(f'元素类型为: {arr1.dtype}')
print(f'元素占用字节数: {arr1.itemsize}')
```

```python
# array()函数 -> 把python中的列表, 封装成 ndarray对象.
arr2 = np.array([1.1, 2.2, 3.3], dtype=np.float32)
print(arr2)
print(type(arr2))
print(f'元素类型为: {arr2.dtype}')
print(f'元素占用字节数: {arr2.itemsize}')
```

## 2.arange()函数

```python
# arange(起始值, 结束值, 步长, dtype=元素类型) -> 类似于Python的range()函数, 生成指定范围的数据, 包左不包右.
arr3 = np.arange(0, 10, 2, dtype=np.int32)
print(arr3)
print(type(arr3))
print(f'元素类型为: {arr3.dtype}')
print(f'元素占用字节数: {arr3.itemsize}')
```

## 3.创建随机数矩阵

```python
# np.random.rand(行, 列) -> 生成0.0 ~ 1.0之间的随机数, 包左不包右.
arr4 = np.random.rand(3, 4)
print(arr4)
print(type(arr4))
print(f'元素类型为: {arr4.dtype}')
print(f'元素占用字节数: {arr4.itemsize}')
```

```python
# np.random.randint(起始值, 结束值, size=(行数, 列数))  -> 生成指定范围内的的随机 整数.
arr5 = np.random.randint(-1, 5, size=(3, 4))
print(arr5)
print(type(arr5))
print(f'元素类型为: {arr5.dtype}')
print(f'元素占用字节数: {arr5.itemsize}')
```

```python
# np.random.uniform(起始值, 结束值, size=(行数, 列数))  -> 生成指定范围内的的随机 小数.
arr6 = np.random.uniform(-1, 5, size=(3, 4))
print(arr6)
print(type(arr6))
print(f'元素类型为: {arr6.dtype}')
print(f'元素占用字节数: {arr6.itemsize}')
```

## 4.创建等比数列

```python
# logspace(起始次幂, 结束次幂, 数字个数, base=基数) -> 生成等比数列, 默认基数为10, 即: 10的起始次幂  ~ 10的结束次幂, 包左包右.  base参数可以修改基数.
# 需求1: 创建长度为5的等比数列, 起始次幂为1, 结束次幂为3
arr7 = np.logspace(1, 3, 5)
print(arr7)
```

```python
# 需求2: 创建长度为10的等比数列, 起始次幂为0, 结束次幂为5, 基数为2
arr8 = np.logspace(0, 5, 10, base=2)
print(arr8)
```

## 5.创建等差数列

```python
# 格式: np.linspace(起始值, 结束值, 元素个数, endpoint=True)    # 生成指定范围内的等差数列, 默认: 包左包右. 
# 需求1: 创建等差数列, 1, 10之间, 10个值.   [1, 10]
arr9 = np.linspace(1, 10, 10)
print(arr9)
```

```python
# 需求2: 创建等差数列, 1, 10之间, 10个值, 包左不包右(前闭后开), [1, 10)
arr10 = np.linspace(1, 10, 10, endpoint=False)
print(arr10)
```

# numpy中的数据类型转换

```python
# 需求: 把 元素类型为 np.float64 -> np.int32
# 1. 创建数组, 指定元素类型为: np.flot64
arr11 = np.zeros((3, 4), dtype=np.float64)
print(arr11)
print(type(arr11))
print(f'元素类型为: {arr11.dtype}')
```

```python
# 2. 把上述的数组元组类型 -> np.int32
arr12 = arr11.astype(np.int32)
print(arr12)
print(type(arr12))
print(f'元素类型为: {arr12.dtype}')
```
