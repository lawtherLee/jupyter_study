# 1. DataFrame数据的增删改查操作

```python
## 1.0 准备动作, 读取数据, 获取df对象.
df1 = pd.read_csv('../data/1960-2019全球GDP数据.csv', encoding='gbk')
df1     # 9930行, 3列

# 基于 df1 拷贝出 df2
df2 = df1.head().copy()
df2     # 5行, 3列
```

## 1.1 增加列.

```python
# 方式1:  df[列名] = 值      会修改原数据.
df2['col1'] = 22                        # 该列会全用 22 来填充.
df2['col2'] = [1, 2, 3, 4, 5]           # 该列会用 列表中的每个值 来对应填充.  细节: 列表长度 = 行数.
df2['col3'] = df2.year * 2              # 还可以用其它列的值 来计算 填充. 
df2
```

```python
# 方式2: df.assign(列名 = 值)    不会修改原数据.
df2.assign(col1 = 22)
df2.assign(col2 = [1, 2, 3, 4, 5])
df2.assign(col3 = df2.year * 2)
```

```python
# assign() 还支持传入 Series对象. 
s1 = pd.Series(['段延庆', '叶二娘', '南海鳄神', '云中鹤'])       # 列表长度 = 行数, 如果不匹配, 用NaN来填充.
df2.assign(col4 = s1)
```

```python
# assign() 还支持 自定义函数. 
def my_fun(df):         # 这里的df就是要被处理的 DataFrame对象.
    return 'hello'

def my_fun1(df):
    return df.index.values * 2  # 返回该df的行索引 * 2

df2.assign(col5 = my_fun)
df2.assign(col6 = my_fun1)
```

```python
# 总结, assign() 可以用来增加列, 不能直接修改原数据, 且 assign()支持同时传入多个列. 
df2 = df2.assign(
    col1 = 22,
    col2 = [1, 2, 3, 4, 5],
    col3 = df2.year * 2,
    col4 = s1,
    col5 = my_fun,
    col6 = my_fun1
)
df2
```

## 1.5 排序相关

```python
# 1. 准备数据.
df1 = pd.read_csv('./data/1960-2019全球GDP数据.csv', encoding='gbk')
df1
```

```python
# 2. 按照行索引排序
df1.sort_index()                    # 默认, 升序.
df1.sort_index(ascending=False)     # ascending=True -> 升序, False -> 降序
```

```python
# 3. 按照 列值 排序.
# 场景1: 按照 单列值 排序.
df1.sort_values('GDP')                      # 默认: 升序
df1.sort_values('GDP', ascending=False)     # 降序.

# 场景2: 按照 多列值 排序, 排序方式一样.
df1.sort_values(['year', 'GDP'])                        # 按照 year列 升序排列, 年份一样再按照 GDP列 升序 排序.
df1.sort_values(['year', 'GDP'], ascending= False)      # 按照 year列 降序排列, 年份一样再按照 GDP列 降序 排序.

# 场景3: 按照 多列值 排序, 排列方式不同, 例如: 按照 year列 降序, GDP升序.
df1.sort_values(['year', 'GDP'], ascending=[False, True])
```

```python
# 4. 排序(排名)函数, rank()函数.
df2 = pd.DataFrame({
    '姓名':['小明', '小美', '小强', '小兰'],
    '成绩':[100, 90, 90, 80]
})
df2
```

```python
# 用rank()函数, 对上述内容做排名. 
df2.rank(ascending=False)                       # 按照成绩, 降序排名. 
df2.rank(ascending=False, method='average')     # 效果同上, 按照成绩, 降序排名.  评分一致, 取平均值
df2.rank(ascending=False, method='min')         # 评分一致, 取最小值.
df2.rank(ascending=False, method='max')         # 评分一致, 取最大值.
#
# # 掌握如下这个, 上述的了解排序原理即可.
df2.rank(ascending=False, method='dense')       # 数值相同, 评分一致.  等价于SQL中的 dense_rank()

# 只对成绩列做排名，不同method效果对比
df2['排名_average'] = df2['成绩'].rank(ascending=False, method='average')
# df2['排名_min'] = df2['成绩'].rank(ascending=False, method='min')
# df2['排名_max'] = df2['成绩'].rank(ascending=False, method='max')
# df2['排名_dense'] = df2['成绩'].rank(ascending=False, method='dense')
df2
```
