## 1.根据行列索引,获取数据

```python
import pandas as pd
# import os
 # 改变相对路径
# os.chdir(r'D:\wordspace\ai_22_work_bj\pandasProject\day02')
```

```python
# 1.准备数据
df = pd.read_csv('../data/stock_day.csv')
df
```

### 1.1直接使用行列索引，先列后行

```python
# 2.根据行列索引 获取数据-->先列后行
#    列         行
df['open']['2018-02-27']
df['open']['2018-02-22']

#      行        列
# df['open']['2018-02-22']  # 报错，必须先列后行

df['high']['2018-02-23']   # 23.37
```

### 1.2 通过loc属性来获取, 先行后列

```python
# 3.通过loc属性来获取，只能通过：行索引+列名的方式来获取
# 格式：df.loc[行索引， 列名]
df.loc['2018-02-23', 'high']
```

```python
# 需求：获取从2018-02-27 - 2018-02-22的open列数据
df.loc['2018-02-27':'2018-02-22', 'open']


# 需求：获取从2018-02-27 - 2018-02-22的open, high, close, low列数据
df.loc['2018-02-27':'2018-02-22', 'open':'low']
df.loc['2018-02-27':'2018-02-22', ['open','high', 'close','low']]

# 需求：获取从2018-02-27 - 2018-02-22的所有列数据
df.loc['2018-02-27':'2018-02-22', :]         # 这里的：代表所有列

# 等价于直接打印df对象，获取所有数据
# df.loc[:, :]
df
```

### 1.3 通过iloc属性来获取, 先行后列, 只能通过：行号+列的编号方式来获取

```python
# 4.通过iloc属性来获取，只能通过：行号+列的编号方式来获取
# 格式: df.iloc[行号， 列的编号]
df.iloc[0, 0] # 0行0列的数据,即你看到的第一行第一列的数据->23.53
```

```python
# 需求：获取从2018-02-27 - 2018-02-22的open列数据
df.iloc[:4, 0]

# 需求：获取从2018-02-27 - 2018-02-22的open, high, close, low列数据
df.iloc[:4, :4]

# 需求：获取从2018-02-27 - 2018-02-22的所有列数据
df.iloc[:4]
df.iloc[:4, :]         # 这里的：代表所有列

# 等价于直接打印df对象，获取所有数据
# df.iloc[:, :]

# 观察效果
df.iloc[1:5:2, 2:5]
```

## 3.DataFrame和Series的排序操作

```python
# 1.查看数据源
df
```

```python
# 2.根据单列值进行排序, 例如根据open列(开盘价)升序排序
df.sort_values('open')
df.sort_values('open', ascending=True)  #  ascending=True-->升序(默认)
df.sort_values('open', ascending=False) #  ascending=False-->降序

```

```python
# 3.根据多列值进行排序
# 场景1:多列值排序规则一样, 例如:按照open升序, open一样,按照high升序
df.sort_values(['open', 'high'])
df.sort_values(['open', 'high'], ascending=True) # 效果同上.
```

```python
# 场景2:多列值排序规则一样, 例如:按照open升序, open一样,按照high降序
df.sort_values(['open', 'high'], ascending=True)
df.sort_values(['open', 'high'], ascending=False)

df.sort_values(['open', 'high'], ascending=[True, False])
```

```python
# 4.根据行索引排序
df.sort_index()                  # 默认升序
df.sort_index(ascending=True)    # 效果同上

df.sort_index(ascending=False)   # 降序
df
```

```python
# 5.无论是上述的根据值排序,还是根据行索引排序, 默认都是按照axis=0进行排序的
# df.sort_values('open', axis='rows')   # 根据open字段, 按照行排序, 默认:升序
df.sort_values('open', axis=0)        # 效果同上.

# df.sort_index('open', axis=1)           # 报错
```

```python
# 6.Series对象的排序, 也有sort_index() 和 sort_values()两个函数, 用法和df对象一样
df.open.sort_index()                 # 按照索引: 升序
df.open.sort_index(ascending=False)  # 按照索引: 降序

df.open.sort_values()                # 按照值: 升序
df.open.sort_values(ascending=False) # 按照值: 降序
```
