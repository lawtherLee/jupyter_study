```python
import pandas as pd
import numpy as np
import matplotlib as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('../data/1960-2019全球GDP数据.csv', encoding='gbk')
df.head()
```

### 绘制中国近年的gdp曲线

```python
china_df = df[df.country == '中国']
china_df.set_index('year', inplace=True) # 参1 索引列名 参2 是否覆盖
# china_df.head()
```

```python
china_df.plot()
```

### 分别查询中美日三国的gdp数据并绘制图形

```python
# 设置索引列
china_df = df[df.country == '中国'].set_index('year')
usa_df = df[df.country == '美国'].set_index('year')
japan_df = df[df.country == '日本'].set_index('year')


# 修改列名
china_df.rename(columns={'GDP': '中国'}, inplace=True)
usa_df.rename(columns={'GDP': '美国'}, inplace=True)
japan_df.rename(columns={'GDP': '日本'}, inplace=True)

# print(china_df)
# print(usa_df)
# print(japan_df)

china_df.中国.plot(legend=True)
usa_df.美国.plot(legend=True)
japan_df.日本.plot(legend=True)
```
