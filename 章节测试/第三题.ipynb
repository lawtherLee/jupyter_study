```python
import pandas as pd
```

```python
data = {
    'city': ['Beijing', 'Shanghai', 'Guangzhou', 'Shenzhen', 'Chengdu', 'Hangzhou'],
    'temperature': [26, 28, 32, 30, 25, 27],
    'humidity': [60, 65, 70, 68, 72, 66]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
温度大于27 = df[df['temperature'] > 27]
```

```python
# 按照湿度从高到低排序
湿度从高到低排序 = df.sort_values(by='humidity', ascending=False)
湿度从高到低排序
```

```python
df['comfort_index'] = df['temperature'] * 0.6 + df['humidity'] * 0.4
```

```python
df['high_humidity'] = df['humidity'] >= 68
df
```

```python
grouped_stats = df.groupby('high_humidity').agg(
    城市 = ('city', list),
    平均温度 = ('temperature', 'mean'),
    平均舒适度 = ('comfort_index', 'mean')
).T
grouped_stats
```
