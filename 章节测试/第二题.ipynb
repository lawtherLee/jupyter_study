```python
import pandas as pd
import numpy as np
from io import StringIO
```

```python
csv_data = """
name,math,english,science
Alice,85,78,92
Bob,90,85,88
Charlie,75,95,85
David,60,,70
Eva,88,76,NaN
"""

df = pd.read_csv(StringIO(csv_data))
print(df)
```

```python
df.isnull().sum()
```

```python
has_na_col = ['math', 'english', 'science']
df[has_na_col] = df[has_na_col].fillna(df[has_na_col].mean())
df
```

```python
df['average'] = df[has_na_col].mean(axis=1).round(2)
df
```
